```python
import pandas as pd
df = pd.read_csv('mexico-real-estate-combined.csv')
print(df.to_string())
```

```python
state_count = df['state'].value_counts().iloc[:10]
print(state_count)
```

```python
property_type_sum = df.groupby('state')['price_usd'].sum().sort_values(ascending=False)
print(property_type_sum)
```

```python
pivot = df.groupby("property_type")["price_usd"].sum().fillna(0)
print(pivot)
```

```python
pivot = df.pivot_table(index='state', columns='property_type', values='price_usd').fillna(0)
print(pivot)
```

```python
descriptive_stats = df.groupby('property_type')['price_usd'].describe()
print(descriptive_stats)

```

```python
#overall descriptive statistics
overall_stats = df['price_usd'].describe()
print(overall_stats)

#descriptive statistics by state
state_stats = df.groupby('state')['price_usd'].describe()
print(state_stats.to_string())
```

```python
pivot = df.pivot_table(index='property_type', columns='state', values='price_usd').fillna(0)
print(pivot.to_string())
```
