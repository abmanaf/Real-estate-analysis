```python
import pandas as pd
df1 = pd.read_csv('mexico-real-estate-1.csv')
df1.dropna(inplace=True)
df1['price_usd'] = df1['price_usd'].str.replace('$', '').str.replace(',', '').astype(float)

print(df1.head())
```

```python
df2 = pd.read_csv('mexico-real-estate-2.csv')
df2.dropna(inplace=True)
df2['price_usd'] = (df2['price_mxn'] / 19).round(2)
df2.drop(columns=['price_mxn'], inplace=True)
df2.info()
print(df2.to_string())
```

```python
df3 = pd.read_csv('mexico-real-estate-3.csv')
df3.dropna(inplace=True)
df3["state"] = df3["place_with_parent_names"].str.split("|", expand=True)[2]
df3.drop(columns=['place_with_parent_names'], inplace=True)
df3[["lat", "lon"]] = df3['lat-lon'].str.split(",", expand=True).astype(float)
df3.drop(columns=['lat-lon'], inplace=True)
df3.info()
print(df3.to_string())
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
# save the combined dataframe to a new CSV file
df.to_csv('mexico-real-estate-combined.csv', index=False)
```
